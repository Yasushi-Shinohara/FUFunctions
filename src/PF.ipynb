```python
import sys
import numpy as np
import math
import matplotlib.pyplot as plt
from modules.constants import *
from modules.Fitting import *
```

```python
# Data construction
N = 20
xs = 0.0
xe = 10.0
x = np.linspace(xs, xe, N)
atrue = 2.0
btrue = -1.0
rseed = 123
np.random.seed(seed = rseed)
randamp = (np.amax(atrue*x + btrue) - np.amin(atrue*x + btrue))/10.0
print("The random noise amplitude: ",randamp)
y = atrue*x + btrue + randamp*np.random.rand(N)   #Linear function is assumed
plt.figure()
plt.xlim(xs, xe)
plt.scatter(x, y, label="Constructed data")
plt.grid()
plt.legend()
plt.show()
```

```python
# Fitting by the new API rather than old one, np.polyfit
norder = 1 #Meaning linear function
mt, _ = Fitting.polynomialfit(x, y, norder)
#
print('mt =', mt)
```

```python
x4fit, y4fit = Fitting.polynomialfunc(x, mt, N4fit = 100)
#
plt.figure()
plt.xlim(np.amin(x4fit), np.amax(x4fit))
plt.scatter(x, y, label="Constructed data")
plt.plot(x4fit, y4fit, label="Fitted curve")
plt.grid()
plt.legend()
plt.show()
```

```python
error = Fitting.polynomialerror(mt, x, y)
#
print('RSS= ',np.sum(error**2))
#
plt.figure()
plt.xlim(xs, xe)
plt.scatter(x, error, label="Error of fitting")
plt.grid()
plt.legend()
plt.show()
```
