```python
import sys
import numpy as np
import math
import matplotlib.pyplot as plt
from modules.FourierAnalysis import *
```

```python
a = 10.0
Nb = 4
Nk = 12
kmin = -pi/a
kmax = pi/a
k = np.linspace(kmin, kmax, Nk, endpoint = False)
ebk = np.zeros([Nb, Nk], dtype='float64')
for ib in range(Nb):
    ebk[ib, :] = 2.0*float(ib) + np.cos((ib+1)*a*k[:])
plt.figure()
plt.xlim(kmin, kmax)
for ib in range(Nb):
    plt.plot(k,ebk[ib, :], label=str(ib)+' th band')
plt.grid()
plt.legend()
plt.show()
```

```python
kfiner, ebkfiner, debkpdk, d2ebkpdk2 = FourierAnalysis.Fourier_interpolation_1D(k, ebk[0,:], Nfiner=10)
plt.figure()
plt.title('Original and interpolated data on finer mech')
plt.xlim(kmin, kmax)
plt.plot(k,ebk[0,:], label='Original')
plt.plot(kfiner,ebkfiner[:],'--',label='Interpolated')
plt.grid()
plt.legend()
plt.show()
#
plt.figure()
plt.xlim(kmin, kmax)
plt.title('First derivative of data ')
plt.plot(kfiner,debkpdk[:])
plt.grid()
plt.show()
#
plt.figure()
plt.xlim(kmin, kmax)
plt.title('Second derivative of data')
plt.plot(kfiner,d2ebkpdk2[:])
plt.grid()
plt.show()
```

```python
kfiner, ebkfiner, debkpdk, d2ebkpdk2 = FourierAnalysis.Fourier_interpolation_1D(k, ebk, Nfiner=10)
plt.figure()
plt.title('Original and interpolated data on finer mech')
plt.xlim(kmin, kmax)
for ib in range(Nb):
    plt.plot(k,ebk[ib,:], label=str(ib)+' th band')
    plt.plot(kfiner,ebkfiner[ib,:],'--')
plt.grid()
plt.legend()
plt.show()
#
plt.figure()
plt.xlim(kmin, kmax)
plt.title('First derivative of data (Normalized+shifted) ')
for ib in range(Nb):
    plt.plot(kfiner,debkpdk[ib,:]/(ib+1)/a+2*ib, label=str(ib)+' th band')
plt.grid()
plt.legend()
plt.show()
#
plt.figure()
plt.xlim(kmin, kmax)
plt.title('Second derivative of data (Normalized+shifted) ')
for ib in range(Nb):
    plt.plot(kfiner,d2ebkpdk2[ib,:]/(ib+1)**2/a**2+2*ib, label=str(ib)+' th band')
plt.grid()
plt.legend()
plt.show()
```
